```python
import requests
import shutil

URL = "https://bilguuntulga.com/wp-content/uploads/2019/08/{index}{part}.jpg"
file_name = "cards/{index}{part}.jpg"

for index in range(1, 41):
    for part in ("a", "b"):
        r = requests.get(URL.format(index=index, part=part), stream=True)
        name = file_name.format(index=index, part=part)

        if r.status_code == 200:
            with open(name, "wb") as f:
                r.raw.decode_content = True
                shutil.copyfileobj(r.raw, f)
            print(f'üü¢ {name}')
        else:
            print(f'‚ùå {name}')
```

```python

def merge_images(card):
    a = Image.open(f'cards/{card}a.jpg')
    b = Image.open(f'cards/{card}b.jpg')

    dst = Image.new('RGB', (a.width + b.width, a.height))
    dst.paste(a, (0, 0))
    dst.paste(b, (a.width, 0))
    return dst.save(f"cards/{card}.jpg")

for card in range(1, 41):
    merge_images(card)
```
